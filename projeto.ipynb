```python
import pandas as pd
df = pd.read_csv('vacinaESAVI.csv')
df.head(2)
```

```python
df.dtypes
```

```python
nome_da_coluna = 'Data da Notificação'
df[nome_da_coluna] = pd.to_datetime(df[nome_da_coluna], format='%d/%m/%Y', errors='coerce')
print(df[nome_da_coluna].dtype)
```

```python
df.isnull().any()
```

```python
# Mostra a contagem de valores vazios por coluna
df.isnull().sum()
```

```python
# Preenche os campos de texto vazios de colunas específicas
df['Município de vacinação'] = df['Município de vacinação'].fillna('Indefinido')
df['Idade Evento'] = df['Idade Evento'].fillna('Não informado')
df['Data da Notificação'] = df['Data da Notificação'].fillna('Não informado')
df['Raça/Cor'] = df['Raça/Cor'].fillna('Não informado')
df['Sexo'] = df['Sexo'].fillna('Não informado')
df['Imunobiológico (vacina)'] = df['Imunobiológico (vacina)'].fillna('Não informado')
df['Dose'] = df['Dose'].fillna('Não informado')
df['Gestante'] = df['Gestante'].fillna('Não informado')
df['Mês de gestação'] = df['Mês de gestação'].fillna('Não informado')
df['Mulher amamentando no momento da vacinação?'] = df['Mulher amamentando no momento da vacinação?'].fillna('Não informado')
df['Tipo de Evento'] = df['Tipo de Evento'].fillna('Não informado')
df['Reação / evento adverso'] = df['Reação / evento adverso'].fillna('Não informado')
df['Classificação de gravidade'] = df['Classificação de gravidade'].fillna('Não informado')
df['Gravidade'] = df['Gravidade'].fillna('Não informado')
df['Desfecho (evolução do caso)'] = df['Desfecho (evolução do caso)'].fillna('Não informado')
df['Doenças (CID10) - Preexistente'] = df['Doenças (CID10) - Preexistente'].fillna('Não informado')
df['Medicamento em uso anterior ou durante a vacinação'] = df['Medicamento em uso anterior ou durante a vacinação'].fillna('Não informado')
df['Nome do Medicamento'] = df['Nome do Medicamento'].fillna('Não informado')
df['Houve atendimento médico?'] = df['Houve atendimento médico?'].fillna('Não informado')
df['Tipo de Atendimento'] = df['Tipo de Atendimento'].fillna('Não informado')


# Mostra o resultado
df.head()
```

```python
#Removendo linhas em branco 
df=df.drop([10,20,182,189,219,242,414,444])
```

```python
# Apaga linhas duplicadas
df = df.drop_duplicates()
```

```python
# Excluindo coluna redundante
df = df.drop(columns=['Gravidade'])
```

```python
# Selecionando apenas as colunas que são do tipo texto
colunas_texto = df.select_dtypes(include=['object', 'string']).columns

for col in colunas_texto:
    # Transformando em maiúsculo todas as String e removendo espaços em branco invisíveis nas pontas
    df[col] = df[col].astype(str).str.upper().str.strip()
```

```python
df.head()
```

```python
# Removendo o padrão "1: " do início dos textos
df['Imunobiológico (vacina)'] = df['Imunobiológico (vacina)'].str.replace(r'^\d+:\s*', '', regex=True)
df['Dose'] = df['Dose'].str.replace(r'^\d+:\s*', '', regex=True)
df['Tipo de Evento'] = df['Tipo de Evento'].str.replace(r'^\d+:\s*', '', regex=True)
df['Reação / evento adverso'] = df['Reação / evento adverso'].str.replace(r'^\d+:\s*', '', regex=True)
df['Classificação de gravidade'] = df['Classificação de gravidade'].str.replace(r'^\d+:\s*', '', regex=True)
df['Desfecho (evolução do caso)'] = df['Desfecho (evolução do caso)'].str.replace(r'^\d+:\s*', '', regex=True)
df['Doenças (CID10) - Preexistente'] = df['Doenças (CID10) - Preexistente'].str.replace(r'^\d+:\s*', '', regex=True)
df['Nome do Medicamento'] = df['Nome do Medicamento'].str.replace(r'^\d+:\s*', '', regex=True)
df['Tipo de Atendimento'] = df['Tipo de Atendimento'].str.replace(r'^\d+:\s*', '', regex=True)

df.head()
```

```python
# Excluindo coluna redundante
df = df.drop(columns=['Gravidade'])
```

```python
df.head()
```

```python
print("Idade mínima:", df['Idade Evento'].min())
print("Idade máxima:", df['Idade Evento'].max())
```

```python
# Exibindo em forma de inteiros 

df['Idade Evento'] = pd.to_numeric(df['Idade Evento'], errors='coerce').astype('Int64')


df['Mês de gestação'] = pd.to_numeric(df['Mês de gestação'], errors='coerce').astype('Int64')

df.head()
```

```python
df.to_csv('vacinaESAVI.csv', index=False, encoding='utf-8')
```
